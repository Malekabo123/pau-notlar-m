```python
from deap import base, tools, creator, algorithms
import random
import numpy as np
from multiprocessing import Pool
```

```python
pool = Pool()
```

```python
mesafe = []
f=open('files/ilmesafe2.csv')
for satir in f.read().split('\n'):
    mesafe.append(list(map(lambda x:0 if x=='' else int(x),satir.split(';'))))
print(len(mesafe[0]))
IND_SIZE = len(mesafe)
creator.create("Fitness", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.Fitness)
```

```python
toolbox = base.Toolbox()
toolbox.register("indices", random.sample, range(IND_SIZE), IND_SIZE)
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
toolbox.population(n=2)
```

```python
def evalTSP(ind):
    route_length = 0
    for i in range(len(ind)-1):
        route_length += mesafe[ind[i]][ind[i+1]]
    route_length += mesafe[ind[0]][ind[-1]]
    return route_length,
```

```python
ind = toolbox.individual()
print(ind)
evalTSP(ind)
```

```python
toolbox.register("evaluate", evalTSP)
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.05)
toolbox.register("mate", tools.cxOrdered)
toolbox.register("select", tools.selTournament, tournsize=5)
```

```python
population = toolbox.population(n=1000)

stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("max", np.max)
stats.register("min", np.min)
stats.register("avg", np.mean)
stats.register("std", np.std)

hof = tools.HallOfFame(5)
```

```python
population, logbook = algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.1, ngen=1000, stats=stats, halloffame=hof)
```
